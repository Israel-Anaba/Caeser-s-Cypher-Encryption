```python
# Caesar's Cipher Encryption

print("Welcome to Cipher service")

def caesar_cipher():
    action = input("Do you want to encrypt or decrypt a message? ").lower()
    message = input("Please enter your message: ").lower()
    key = int(input("Please enter the key (1-26): "))
    
    # Encryption
    if action == 'encrypt':
        cipher = ''
        for char in message:
            if char == ' ':
                cipher = cipher + char
            elif char.isupper():
                cipher = cipher + chr((ord(char)+key-65)%26 + 65)
            else:
                cipher = cipher + chr((ord(char)+key-97)%26 + 97)
        print("The encrypted message is:", cipher)
    
    # Decryption
    else:
        decipher = ''
        for char in message:
            if char == ' ':
                decipher = decipher + char
            elif char.isupper():
                decipher = decipher + chr((ord(char)-key-65)%26 + 65)
            else:
                decipher = decipher + chr((ord(char)-key-97)%26 + 97)
        print("The decrypted message is:", decipher)

caesar_cipher()
```

```python
# Caesar's Cipher Encryption

print("Welcome to the Cipher Service")

# Dictionary mapping plaintext to ciphertext
cipher = {
    'A':'X', 'B':'Y', 'C':'Z', 'D':'A', 'E':'B', 'F':'C', 'G':'D', 'H':'E',
    'I':'F', 'J':'G', 'K':'H', 'L':'I', 'M':'J', 'N':'K', 'O':'L', 'P':'M',
    'Q':'N', 'R':'O', 'S':'P', 'T':'Q', 'U':'R', 'V':'S', 'W':'T', 'X':'U',
    'Y':'V', 'Z':'W'
}


plaintext = input('Do you want to encrypt or decrypt your message?  ').lower()

def encrypt(plaintext):
    ciphertext = ""
    if plaintext == 'e':
        message = input('What is the message you want to encrypt? ').lower()
        ciphertext = encrypt(message)
        
    for char in plaintext:
        if char.isalpha():
            ciphertext += cipher[char.upper()]
        else:
            ciphertext += char
    return ciphertext



def decrypt(ciphertext):
    plaintext = ""
    for char in ciphertext:
        if char.isalpha():
            for key, value in cipher.items():
                if value == char.upper():
                    plaintext += key
        else:
            plaintext += char

    return plaintext


if plaintext == 'e':
    ciphertext = encrypt(plaintext)
    print(ciphertext)

elif plaintext == 'd':
    ciphertext = input('What is the message you want to decrypt? ').lower()
    decrypted = decrypt(ciphertext)
    print(decrypted)
```
